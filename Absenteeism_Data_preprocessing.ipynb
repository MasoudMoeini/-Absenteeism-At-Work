# Data Preprocessing, Data Analysing

```python
import pandas as pd
raw_csv_data=pd.read_csv('Absenteeism_data.csv')
raw_csv_data.head()
```

To keep the original data secure we use another variable to start doing some modification in our raw data

```python
df=raw_csv_data.copy()
df.info()
```

Choosing Model and Prediction: The grund truth is "Absenteeism Time in Hours "
which is a dependent variable and all of the other columns are potentially independent variables as an input<br>
## First step: Removing Irrelevant input- feature data that has no effect on our prediction/Feature selection

```python
df=df.drop(['ID'],axis=1) #axis 0 points towards rows, and axis=1 hints towards columns
df['Reason for Absence'].min()
df['Reason for Absence'].max()
df['Reason for Absence'].unique()#We want to extract a list containing distinct values only
```

```python
len(df['Reason for Absence'].unique())
```

THe numbers in this column are actually categorical numbers which inherently represent some reasons from the list but and there is no mathematically values/comparison between them e.g 25 is not greater than 12 here, we just prefer to use numbers instead of characters and text<br>
How to add numeric values to our categorical numbers for the purpose of quantitative analysis ??<br>
so the solution is Dummy variables and one hot coding, if from the list the reason of absence is number 10 ,
we make it 1 and the other will be assigned 0 , we use .get_dummies()

```python
reason_columns=pd.get_dummies(df['Reason for Absence'])
reason_columns.head()
```

```python
reason_columns['check']=reason_columns.sum(axis=1)
reason_columns.head()
```

```python
reason_columns['check'].sum(axis=0)
```

```python
reason_columns=reason_columns.drop(['check'],axis=1)
reason_columns=pd.get_dummies(df['Reason for Absence'],drop_first=True)
df.columns.values
```

```python
df=df.drop(['Reason for Absence'],axis=1)
```

If we use dummy variables then the problem would be numerous columns, we are interested to reduce this huge number of columns therefore we use grouping and classifying technique to summerize some of these columns

```python
reason_columns.loc[:,1:28].head()
```

```python
reason_type_1=reason_columns.loc[:,1:14].max(axis=1)
reason_type_2=reason_columns.loc[:,15:17].max(axis=1)
reason_type_3=reason_columns.loc[:,18:21].max(axis=1)
reason_type_4=reason_columns.loc[:,21:].max(axis=1)
```

```python
df=pd.concat([df,reason_type_1,reason_type_2,reason_type_3,reason_type_4],axis=1)
df.head()
```

```python
column_names=['Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours', 'Reason_1','Reason_2','Reason_3','Reason_4']
```

```python
df.columns=column_names
df.head()
```

# Reorder Columns

```python
column_names_reordered=['Reason_1','Reason_2','Reason_3','Reason_4','Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours']
df=df[column_names_reordered]
df.head()
```

Create Checkpoint<br>
Create a copy of the current state of the df data frame

```python
df_reason_mod=df.copy()
df_reason_mod['Date'][0]
```

```python
type(df_reason_mod['Date'][0])
```

# Convert date values to Timestamp

```python
df_reason_mod['Date']=pd.to_datetime(df_reason_mod['Date'],format='%d/%m/%Y')
df_reason_mod['Date'].head()
```

```python
df_reason_mod.info()
```

```python
df_reason_mod['Date'][0]
```

```python
df_reason_mod['Date'][0].month
```

```python
len(df_reason_mod)
df_reason_mod.shape
```

```python
list_months=[]
for i in range(df_reason_mod.shape[0]):
    list_months.append(df_reason_mod['Date'][i].month)
```

# Adding new Column to the data frame

```python
df_reason_mod['Month value']=list_months
df_reason_mod.head()
```

```python
df_reason_mod['Date'][0].weekday()
```

```python
def date_to_weekday(date_value):
    return date_value.weekday()
```

If we want to repeatedly apply some changes to the elements of list we use .apply()

```python
df_reason_mod['Day of the week']=df_reason_mod['Date'].apply(date_to_weekday)
df_reason_mod.head()
```

```python
df_reason_mod.drop(['Date'],axis=1)
df_reason_mod.head()#We get ride of column Date 
```

Reordering columns

```python
column_names_reordered_second=['Reason_1','Reason_2','Reason_3','Reason_4','Month value','Day of the week', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours']
df_reason_mod=df_reason_mod[column_names_reordered_second]
df_reason_mod.head()
```

```python
df_reason_date_mod=df_reason_mod.copy()
#to avoid any modification in original preprocessed data until this step
```

Education represents categorical data(1 high school,2 graduate,3 post graduate,4 a master or doctor)<br>Therefore we transform education into dummy variable.<br>Children, Pets arer nummeric values and has a numeric sense no need to transform

```python
df_reason_date_mod['Education'].unique()
```

```python
df_reason_date_mod['Education'].value_counts()
```

Whereas 3,2,4 cover low percentage group of target people for this prediction we consider them all in one class for dummy variable

```python
df_reason_date_mod['Education']=df_reason_date_mod['Education'].map({1:0,2:1,3:1,4:0})
# becareful to use all existing values in map method otherwise we get some Nan data
df_reason_date_mod['Education'].unique()
```

```python
df_reason_date_mod['Education'].value_counts()
```

# Final checkpoint

```python
df_preprocessed=df_reason_date_mod.copy()
df_preprocessed.head()
```
